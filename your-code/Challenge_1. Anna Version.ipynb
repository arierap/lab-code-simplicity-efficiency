```python
## TEST 2 - FINALE PERO NO ESTA LIMITAT LA OPCIO DE CALCULATION OPTION!!!

# Program make a simple calculator
# This function adds two numbers 

def add(x, y):
   return x + y
# This function subtracts two numbers 
def subtract(x, y):
   return x - y

print('''Welcome to this calculator!
It can add and subtract whole numbers from zero to five!''')

print("Select operation.")
print("1.Add")
print("2.Subtract")


# Take input from the user - what calculation
calculation_choice = input("Enter choice 1/2: ")


# Take input from the user - what numbers

num1 = int(input("Please choose your FIRST number from 0 to 5: "))

while 0 > num1 or 5 < num1:
    try:
        num1 = int(input("! Remember, it has to be a number between (0 to 5) : "))
    except ValueError:
        print ("That wasn't an integer :(")
    
num2 = int(input("Please choose your SECOND number from 0 to 5: "))

while 0 > num2 or 5 < num2:
    try:
        num2 = int(input("! Remember, it has to be a number between (0 to 5) : "))
    except ValueError:
        print ("That wasn't an integer :(")
        

if calculation_choice == '1':
   print(num1,"+",num2,"=", add(num1,num2))
elif calculation_choice == '2':
   print(num1,"-",num2,"=", subtract(num1,num2))
else:
   print("Invalid input")
```
